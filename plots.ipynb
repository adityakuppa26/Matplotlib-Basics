Line Plot :

With matplotlib, you can create a bunch of different plots in Python. The most basic plot is the line plot. A general recipe is given here.



```python
year=[2001,2003,2005,2007,2009,2011,2013,2015,2017,2019,2021,2023]
pop=[3.23,4.23,5.31,6.22,6.59,300,234,33,22,1,45,67]

# Print the last item from year and pop
print(year[len(year)-1])
print(pop[len(pop)-1])


# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Make a line plot: year on the x-axis, pop on the y-axis
plt.plot(year,pop)

# Display the plot with plt.show()
plt.show()
```

Scatter Plot :

When you have a time scale along the horizontal axis, the line plot is your friend. But in many other cases, when you're trying to assess if there's a correlation between two variables, for example, the scatter plot is the better choice.

```python
# Make a scatter plot: year on the x-axis, pop on the y-axis
plt.scatter(year,pop)

# Display the plot with plt.show()
plt.show()
```

Histograms :

Used to study the distribution of the data in a dataset.

```python
ages=[20,30,40,50,60,45,56,67,34,56,12,78,87,45,47,32,23,56]
plt.hist(ages)            # default bins=10
plt.show()
```

```python
plt.hist(ages,bins=5)
plt.show()
```

Customizations :

xlabel() , ylabel(), title() , yticks(), xticks(), size

```python
plt.scatter(year,pop)
plt.xlabel('year')
plt.ylabel('population in billions')
plt.title('year vs pop')
plt.yticks([3.5,4,4.5,5.0,5.5,6.0,6.5],['3.5 B','4.0 B','4.5 B','5.0 B','5.5 B','6.0 B','6.5 B'])
plt.show()
```

```python
import numpy as np
np_pop=np.array(pop)
plt.xlabel('year')
plt.ylabel('population in billions')
plt.title('year vs pop')
plt.scatter(year,pop,s=np_pop)       # size of dots as per the population
plt.show()
```

```python
plt.scatter(year,pop,s=np_pop,alpha=0.4)    # alpha for transparency
plt.xlabel('year')
plt.ylabel('population in billions')
plt.title('year vs pop')
plt.show()
```

```python
plt.scatter(year,pop,s=np_pop,alpha=0.4)
plt.xlabel('year')
plt.ylabel('population in billions')
plt.title('year vs pop')
plt.grid(True)           # for grid lines
plt.text(2014,50,'India')    # text in graphs
plt.show()
```
